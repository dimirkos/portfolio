На первый взгляд в данных нет аномалий, пропуски и дубликаты отсутствуют. Предварительно нужно:

### Продуктовые метрики

#### Расчет количества уникальных пользователей в день, неделю, месяц (DAU, WAU, MAU)

```python
#создаем столбцы с днем, неделей, месяцем и годом начала сессии пользователя на сайте
visits_log['session_day'] = visits_log['start_ts'].dt.date
visits_log['session_week'] = visits_log['start_ts'].dt.week
visits_log['session_month'] = visits_log['start_ts'].dt.month
visits_log['session_year'] = visits_log['start_ts'].dt.year
```

```python
dau_total = visits_log.groupby(['session_day']).agg({'uid':'nunique'})
print('Среднее количество уникальных пользователей в день:', int(dau_total.mean()))
dau_total.plot(figsize = (16,8), grid = True, legend = False)
plt.title('Количество уникальных пользователей сайта в день', size = 20)
plt.xlabel('День', size = 12)
plt.ylabel('Количество уникальных пользователей', size = 12)
plt.show()
```

```python
wau_total = visits_log.groupby(['session_year', 'session_week']).agg({'uid':'nunique'})
print('Среднее количество уникальных пользователей в неделю:', int(wau_total.mean()))
wau_total.plot(figsize = (16,8), grid = True, legend = False)
plt.title('Количество уникальных пользователей сайта в неделю', size = 20)
plt.xlabel('Неделя', size = 12)
plt.ylabel('Количество уникальных пользователей', size = 12)
plt.show()
```

```python
mau_total = visits_log.groupby(['session_year', 'session_month']).agg({'uid':'nunique'})
print('Среднее количество уникальных пользователей в месяц:', int(mau_total.mean()))
mau_total.plot(figsize = (16,8), grid = True, legend = False)
plt.title('Количество уникальных пользователей сайта в месяц', size = 20)
plt.xlabel('Месяц', size = 12)
plt.ylabel('Количество уникальных пользователей', size = 12)
plt.show()
```

Заметно, как вырастают показатели к Новому году. После Нового года, в 2018 году у сайта пользователей больше чем летом 2017.

```python
sticky_wau = float(dau_total.mean()/wau_total.mean())

sticky_mau = float(dau_total.mean()/mau_total.mean())

print('Степень лояльности и вовлеченности аудитории за неделю {:.0%}'.format(sticky_wau))
print('Степень лояльности и вовлеченности аудитории за месяц {:.0%}'.format(sticky_mau))
```

#### Расчет "коэффициента удержания" (Retention Rate). Как часто люди возвращаются на сайт.

```python
first_visit = visits_log.groupby(['uid'])['start_ts'].min()
first_visit.name = 'first_visit'
visits = visits_log.join(first_visit, on = 'uid')
```

```python
visits['first_visit_month'] = visits['first_visit'].astype('datetime64[M]')
visits['start_ts_month'] = visits['start_ts'].astype('datetime64[M]')
visits['lifetime'] = ((visits['start_ts_month'] - visits['first_visit_month'])
                         / np.timedelta64(1, 'M')).round().astype('int')
```

```python
cohorts_visits = visits.groupby(['first_visit_month','lifetime']).agg({'uid':'nunique'}).reset_index()
users_count = cohorts_visits[cohorts_visits['lifetime'] == 0][['first_visit_month','uid']]
users_count.columns = ['first_visit_month', 'cohort_users']
```

```python
cohorts_visits = cohorts_visits.merge(users_count, on = 'first_visit_month')
cohorts_visits['retention'] = cohorts_visits['uid']/cohorts_visits['cohort_users']
cohorts_visits
```

```python
retention_pivot = cohorts_visits.pivot_table(index='first_visit_month',columns='lifetime',values='retention',aggfunc='sum')
retention_pivot.index = retention_pivot.index.strftime("%m-%Y")

sns.set(style='whitegrid')
plt.figure(figsize=(16, 10))
sns.heatmap(retention_pivot, annot=True, fmt='.2%', linewidths=0.2, vmax = 0.1, linecolor='gray', cmap= 'rocket')
plt.title('Коэффициент удержания по когортам (Retention Rate)')
plt.ylabel(' ')
plt.xlabel(' ')
plt.show()
```

```python
print('Средний коэффицент удержания на второй месяц равен {:.2%}'.format(retention_pivot[1].mean().round(4)))
```

Самый высокий коэффициент удержания приходится на первую когорту. В последующих когортах Retention Rate снижается. Мало кто ходит стабильно каждый месяц на мероприятия. Также заметно небольшое увеличение 1ой и 2ой когорт на 4 месяц жизни когорты, возможно пользователям этих когорт прислали промокоды на скидку.

### Метрики электронной коммерции

#### Cколько времени в среднем проходит с момента первого посещения сайта до совершения покупки

```python
first_order = orders_log.groupby('uid')['buy_ts'].min()
first_order.name = 'first_order'
visits = visits.merge(first_order,on='uid')

visits['minute_of_first_order'] = ((visits['first_order'] - visits['first_visit'])/np.timedelta64(1, 'm')).round(1)
```

```python
print('Количество сессий с отрицательными минутами с момента первого визита до первого заказа',
      visits['minute_of_first_order'][visits['minute_of_first_order'] < 0].count())
print('Количество сессий нулевых сессий с момента первого визита до первого заказа', 
      visits['minute_of_first_order'][visits['minute_of_first_order'] == 0].count())
```

Стоит сказать разработчикам, что есть ошибка - заказ выполняется раньше, чем на сайт заходят, или же время первого посещения сайта и время первого заказа одинаковое. Сделаем срез.

```python
visits = visits[visits['minute_of_first_order'] > 0]
```

```python
print('Мода количества минут, за которые пользователи делают первый заказ', visits['minute_of_first_order'].mode()[0])
visits['minute_of_first_order'].plot(kind = 'hist', bins = 50, alpha = 0.5, figsize = (16, 8), range = (0, 10000))
device_session = visits.groupby('device')['minute_of_first_order'].plot(kind='hist', bins=50, alpha=0.5, range = (0, 10000))
plt.title('Гистограмма количества минут от первого визита до первого заказа')
plt.xlabel('Количество минут')
plt.ylabel('Количество пользователей')
plt.legend()
plt.show()
```

Выглядит очень странно, самое популярное значение 1 минута. Для того, чтобы купить за одну минуту билеты нужно быть хорошо знакомым с сайтом, точно знать на какое событие идешь и какие места выбрать. 
В общем стоит опять обратиться к разработчикам и спросить, что не так. Такое впечатление как будто роботы покупали, либо это значение в одну минуту считалось через API Яндекс Афиши(допустим на том же Кинопоиске), поставлено по умолчанию(также это может относиться к отрицательным и нулевым значениям).

#### Расчет среднего количества покупок на одного покупателя за определённый период

```python
orders_log.head()
```

```python
orders = orders_log.join(first_order, on = 'uid')
orders['first_order_month'] = orders['first_order'].astype('datetime64[M]')
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
orders
```

```python
n_buyers = orders.groupby('first_order_month').agg({'uid':'nunique'}).reset_index()
n_buyers.columns = ['first_order_month', 'n_buyers']
n_buyers
```

```python
cohorts_orders = orders.groupby(['first_order_month','order_month']).agg({'uid': 'count'}).reset_index()
cohorts_orders.head()
```

```python
report_orders = pd.merge(n_buyers, cohorts_orders, on='first_order_month')
report_orders.head()
```

```python
report_orders['age'] = (report_orders['order_month'] - report_orders['first_order_month']) / np.timedelta64(1, 'M')
report_orders['age'] = report_orders['age'].round().astype('int')
report_orders['order_per_user'] = report_orders['uid'] / report_orders['n_buyers']

report_orders['first_order_month'] = report_orders['first_order_month'].dt.strftime('%Y-%m')  
pivot_report = report_orders.pivot_table(index='first_order_month', columns='age', values='order_per_user', aggfunc='mean').cumsum(axis=1)

sns.set(style='white')
plt.figure(figsize=(16, 9))
plt.title('Среднее количество покупок одного пользователя по месяцам')
sns.heatmap(pivot_report, annot=True, fmt='.2f', linewidths=0.2, linecolor='black', cmap= 'YlGnBu')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
```

#### Как меняется LTV на покупателя по когортам

**LTV** (от англ. lifetime value) — «пожизненная» ценность клиента. Это — итоговая сумма денег, которую один клиент приносит в среднем со всех своих покупок.

```python
cohorts_orders.head()
```

```python
orders.head()
```

```python
u_buyers = orders.groupby('first_order_month').agg({'uid':'nunique'}).reset_index()
u_buyers.columns = ['first_order_month', 'n_buyers']
u_buyers
```

```python
cohorts = orders.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()
cohorts.head()
```

```python
report = pd.merge(u_buyers, cohorts, on='first_order_month')
report.head()
```

```python
report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
report['ltv'] = report['revenue'] / report['n_buyers']

report['first_order_month'] = report['first_order_month'].dt.strftime('%Y-%m')  
pivot_report = report.pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='mean').cumsum(axis=1)

sns.set(style='white')
plt.figure(figsize=(16, 9))
plt.title('LTV')
sns.heatmap(pivot_report, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
```

Самыми прибыльными считаются первая и четвертая когорты. У четвертой произошел некоторый скачок на третий месяц жизни.

#### Расчет ROMI по когортам в разрезе источников

**ROMI** (Return on Marketing Investment) — коэффициент возврата маркетинговых инвестиций. Эта метрика показывает рентабельность затрат на маркетинг

```python
report['first_order_month'] = report['first_order_month'].astype('datetime64[M]')
romi = pd.merge(report, monthly_costs, left_on='first_order_month', right_on='month')
romi['cac'] = romi['costs'] / romi['n_buyers']
romi['romi'] = romi['ltv'] / romi['cac']
romi_pivot = romi.pivot_table(index='first_order_month', columns='age', values='romi', aggfunc='mean')

romi_pivot = romi_pivot.cumsum(axis=1).round(2)
romi_pivot.index = romi_pivot.index.strftime("%Y-%m")
romi
```

```python
sns.set(style='white')
plt.figure(figsize=(16, 9))
plt.title('ROMI')
sns.heatmap(romi_pivot, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.ylabel(' ')
plt.xlabel(' ')
plt.show()
```

Окупились только 1ая и 4ая когорты.

```python
romi_by_source = cac.merge(revenue_by_source, left_on='source_id', right_on='first_uid_source')
romi_by_source['romi']=romi_by_source['revenue']/romi_by_source['costs']
romi_by_source
```

```python
output = romi_by_source.sort_values(by='romi', ascending=False).reset_index()
output.plot(y='romi', x='source_id', kind='bar', figsize=(16, 9), title='ROMI по источникам', legend=False, grid = True)
plt.xlabel('Источник')
plt.ylabel('Значение ROMI')
plt.show()
```
