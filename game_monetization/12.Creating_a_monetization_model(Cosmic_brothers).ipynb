На графиках заметно увеличение количества пользователей с 4 мая 2020 года по 10 мая 2020 года, далее происходит резкий спад игроков, возможно связанный из-за начала рабочей недели.

```python
dau_total = dau_user.mean()
wau_total = wau_user.mean()
sticky_wau = dau_total / wau_total
print('Среднее количество пользователей в день за весь период', dau_total[0].round(1))
print('Среднее количество пользователей в неделю за весь период', wau_total[0].round(1))
print('Sticky factor за неделю составляет порядка {0:.0%}'.format(sticky_wau[0].round(2)))
```

###  Время прохождения уровня

Для того, чтобы посчитать время прохождения уровня, нужно:
1. Узнать дату первого захода в игру.
2. Дата завершения уровня.
3. Объединить эти даты в общую таблицу и вычислить время прохождения уровня.

```python
#узнаем дату первого захода в игру
start_date = game_actions.groupby('user_id').agg({'event_datetime':['min','max']}).reset_index()
start_date.columns = ['user_id', 'start_date', 'end_date']

#соединим датасеты 
game_actions_win = game_actions.merge(start_date, on = 'user_id')

#удалим нерелевантные значения 
game_actions_win = game_actions_win[game_actions_win['finish_type'] != 'not_finish']

#дата завершения уровня
game_actions_win['winner_time'] = game_actions_win['end_date'] - game_actions_win['start_date'] 
```

## Проверка гипотез.

### Завершение уровня различается в зависимости от способа прохождения?

```python
#посмотрим на датасет с игроками, которые завершили уровень
game_actions_win.head()
```

```python
game_actions_win['winner_time_hours'] = game_actions_win['winner_time'] / np.timedelta64(1, 'h')


plt.figure(figsize = (10, 8))
sns.distplot(game_actions_win.query('finish_type == "project"')['winner_time_hours'],
             color = "skyblue", label = "Реализация проекта", kde = True)
sns.distplot(game_actions_win.query('finish_type == "pvp"')['winner_time_hours'], 
             color = "red", label = "Победа над первым врагом", kde=True)


plt.xlabel('Количество часов')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8, 6))
sns.boxplot(game_actions_win['winner_time_hours'])

plt.title('Диаграмма размаха количества часов для прохождения уровня')
plt.xlabel('Количество часов')
plt.show()
```

```python
#удалим выбросы 
game_actions_win = game_actions_win[(game_actions_win['winner_time_hours'] < 
                                     game_actions_win['winner_time_hours'].quantile(0.99)) &
                                   (game_actions_win['winner_time_hours'] > 
                                     game_actions_win['winner_time_hours'].quantile(0.01))]
```

Сформулируем нулевую и альтернативную гипотезы:

- H0: Время прохождения уровня не зависит от способа прохождения

- Н1: Время прохождения уровня различно

```python
alpha = .05 # критический уровень статистической значимости

results = st.ttest_ind(game_actions_win.query('event == "project"')['winner_time_hours'],
                       game_actions_win.query('event == "finished_stage_1"')['winner_time_hours'])

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Существуют статистически значимые различия между временем прохождея уровня через победу над первым врагом и реализацией проекта. Значит скорость прохождения зависит от способа прохождения, об этом стоит сказать разработчикам игры.

## Модель монетизации

- немного слов о модели прогнозирования оттока клиентов

Так как в датасете использованные данные по игре пользователей на первом уровне, то сложно судить об оттоке пользователей, а также в данных недостаточно признаков для построения качественной модели. В целом логично сказать, чем меньше игрок строит зданий, тем больше вероятность, что он уйдет в отток. Вдобавок, по диаграмме размаха количества часов для прохождения уровня в пункте 4.1 видно, что большинство пользователей проходят уровень в течение 21 дня, соответственно пользователь, чье прохождение первого уровня дольше 21 дня, с большей вероятностью уйдет в отток.

- модель монетизации

Учитывая среднее значение количества построенных зданий для всех игроков, равное 10 и стоимости привлечения одного пользователя от 0.40 до 0.79 ед. можно сформулировать модель монетизации:

- показ рекламы на экране с выбором типа объекта для постройки проходит каждый раз и стоимость должна составлять 0.79/10 = 0.08 без учета оттока пользователей от рекламы;

- показ рекламы на экране с выбором типа объекта для постройки проходит через раз и стоимость должна составлять 0.57*2/10 = 0.16 без учета оттока пользователей от рекламы;

Для прогноза оттока пользователей от рекламы нужно в будущем исследовать данные, связанные уже, непосредственно, со свойствами рекламы.

Для прогноза LTV и ROMI по источникам будем использовать первую модель. 

### Прогноз LTV

```python
#создаем столбец с доходом от показа рекламы
def add_revenue(data):
    if data == 'building':
        return 0.08
    else:
        return 0
    
game_actions['revenue'] = game_actions['event'].apply(add_revenue)  
```

```python
#оставим в таблице значения, в которых был показ рекламы
actions = game_actions[game_actions['revenue'] != 0]
actions.head()
```

```python
#первый показ рекламы
first_actions = actions.groupby('user_id')['session_date'].min()
first_actions.name = 'first_actions'
actions = actions.merge(first_actions, on = 'user_id')
actions.head()
```

```python
actions = actions.merge(user_source, on = 'user_id')
actions.head()
```

```python
cohorts_actions = actions.groupby(['first_actions','session_date']).agg({'user_id': 'count'}).reset_index()
cohorts_actions.head()
```

```python
u_actions = actions.groupby('first_actions').agg({'user_id':'nunique'}).reset_index()
u_actions.columns = ['first_actions', 'n_actions']
u_actions
```

```python
cohorts = actions.groupby(['first_actions','session_date']).agg({'revenue': 'sum'}).reset_index()
cohorts.head()
```

```python
report = pd.merge(u_actions, cohorts, on = 'first_actions')
report.head()
```

```python
report['age'] = (report['session_date'] - report['first_actions']) / np.timedelta64(1, 'D')
report['age'] = report['age'].round().astype('int')
report['ltv'] = report['revenue'] / report['n_actions']


pivot_report = report.pivot_table(index='first_actions', columns='age', values='ltv', aggfunc='mean').cumsum(axis=1)

sns.set(style='white')
plt.figure(figsize=(30, 8))
plt.title('LTV')
sns.heatmap(pivot_report, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
```

Каждый день LTV растет на 0.03 - 0.08. На 19 день рост прекращается у первой когорты, у второй на два дня раньше. Показатели всех когорт перестают сильно изменяться на 12 день. Чем младше когорта, тем LTV меньше растет. 

### Прогноз ROMI по когортам

```python
#сгруппируем данные
actions_grouped = actions.groupby(['source','first_actions', 'session_date']).agg({'user_id': 'nunique', 
                                                                                   'event': 'count'}).reset_index()

actions_grouped.columns = ['source', 'first_actions', 'session_date', 'users_number', 'events_number']
```

```python
ad_costs['day'] = ad_costs['day'] + pd.Timedelta(days=1)
```

```python
actions_grouped['cohort_lifetime'] = ((actions_grouped['session_date']
                                        - actions_grouped['first_actions'])/ np.timedelta64(1, 'D')
)

actions_grouped['cohort_lifetime'] = actions_grouped['cohort_lifetime'].round().astype('int')

romi = pd.merge(actions_grouped, ad_costs, on = 'source')
```

```python
#создадим столбцы с cac, ltv и romi
romi['cac'] = romi['cost'] / romi['users_number']
romi['ltv'] = (romi['events_number'] * 0.08) / romi['users_number']
romi['romi'] = romi['ltv'] / romi['cac']
```

```python
pivot_romi = romi.pivot_table(index = 'first_actions', columns = 'cohort_lifetime', values = 'romi', aggfunc='mean')

pivot_romi = pivot_romi.cumsum(axis=1).round(2)
```

```python
sns.set(style='white')
plt.figure(figsize=(23, 4))
plt.title('ROMI', size=14)
ax=sns.heatmap(pivot_romi, annot=True, fmt='.1f', linewidths=.01, linecolor='gray', cmap="cubehelix")
ax.set_ylabel(' ')
ax.set_xlabel(' ')
plt.show()
```

Не окупаются только две последние когорты.

- Существуют статистически значимые различия между среднем количеством зданий пользователей, которые завершают уровень через реализацию проекта и пользователей победивших первого врага.

**Была предложена модель монетизации:**

*Показ рекламы на экране с выбором типа объекта для постройки проходит каждый раз и стоимость должна составлять 0.08 без учета оттока пользователей от рекламы*

Произведен расчет LTV, ROMI по когортам и ROMI по источникам рекламы для такой модели монетизации.

- LTV: Каждый день LTV растет на 0.03 - 0.08. На 19 день рост прекращается у первой когорты, у второй на два дня раньше. Показатели всех когорт перестают сильно изменяться на 12 день. Чем младше когорта, тем LTV меньше растет;
- ROMI по когортам: Не окупаются только две последние когорты(9 и 10 мая)
- ROMI по источникам рекламы: лучше всех себя ведет ютуб, а вот фэйсбук совсем немного не окупается.

**Рекомендации**

1. Для привлечения потока пользователей стоит чаще использовать ютуб и Яндекс.Директ.
2. Минимальная стоимость показа рекламы должна составлять не менее 0.08 ед.
3. Стоит поговорить с разработчиками о здании исследовательского центра, возможно он очень сложный, поэтому его реже выбирают; а также стоит предложить снизить сложность прохождения уровня через реализацию проекта, т.к. время на него тратится в среднем больше, чем на победу над первым врагом.
